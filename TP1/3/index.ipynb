```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
# line chart
plt.figure(figsize=(10,5))
plt.plot(data['Negara'], data['Jumlah Penduduk'])
plt.xlabel('Negara')
plt.ylabel('Jumlah Penduduk')
plt.title('Jumlah Penduduk di ASEAN')
plt.show()
```
