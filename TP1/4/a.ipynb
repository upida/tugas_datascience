```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
# simulasi 2000 kali lemparan koin Xâˆ¼Bernoulli(0.55) dengan bernoulli.rvs()

n = 2000

x = bernoulli.rvs(0.55, size=n)

heads_count = np.sum(x)
tails_count = n - heads_count

print(f'heads: {heads_count}, tails: {tails_count}')
```

```python
plt.hist(x)
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```
