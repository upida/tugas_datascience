```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
n = 2000
p = 0.5
coin_flips_rvs = bernoulli.rvs(p, size=n)

average_rvs = np.cumsum(coin_flips_rvs) / np.arange(1, n + 1)
average_tails_rvs = 1 - average_rvs

plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, n + 1), average_rvs, label='Rata-rata Kemunculan Heads (1)', color='blue')
plt.plot(np.arange(1, n + 1), average_tails_rvs, label='Rata-rata Kemunculan Tails (0)', color='orange')

plt.title('Perubahan Rata-rata Kemunculan Heads dan Tails setelah Lemparan (p=0.5)')
plt.xlabel('Jumlah Lemparan')
plt.ylabel('Rata-rata Kemunculan')
plt.legend()

plt.grid(True)
plt.show()
```
